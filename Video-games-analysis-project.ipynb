## 3.7 Cross-Platform Comparison

Compare sales performance of games across different platforms:

```python
# Find games released on multiple platforms
def find_multiplatform_games(vgames_relevant):
    platform_counts = vgames_relevant.groupby('name')['platform'].nunique()
    multiplatform_games = platform_counts[platform_counts > 1].index.tolist()
    
    # Group by 'name' and aggregate 'platform' and 'total_sales'
    result_vgames = vgames_relevant[vgames_relevant['name'].isin(multiplatform_games)].groupby('name').agg({
        'platform': lambda x: list(x.unique()),
        'total_sales': 'sum'
    }).reset_index()
    
    return result_vgames

multiplatform_games = find_multiplatform_games(vgames)
display(multiplatform_games.sample(10))
```

```python
# Compare sales across platforms for these games
# Your code here to analyze and visualize cross-platform performance

exploded_vgames = multiplatform_games.explode('platform')
platform_sales = exploded_vgames.groupby('platform')['total_sales'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='platform', y='total_sales', data=platform_sales, palette='plasma')
plt.title('Sales Comparison for Multiplatform Games')
plt.xlabel('Platform')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
       
   
```

<div class="alert alert-info" style="border-radius: 15px; box-shadow: 4px 4px 4px; border: 1px solid ">
<b> Student's comment: </b>

# Step 5 : Hypothesis Tests

—Average user ratings of the Xbox One and PC platforms are the same.

—Average user ratings for the Action and Sports genres are different.

Set the *alpha* threshold value yourself.

Explain:

—How you formulated the null and alternative hypotheses

—What criteria you used to test the hypotheses~~,~~ and why


<div class="alert alert-info" style="border-radius: 15px; box-shadow: 4px 4px 4px; border: 1px solid ">
    <b>   Student's comment: </b> 

**1.** I used filtering acording to the tasks, first by platform (Xbox one and PC) and then by genre (Action and Sports) as the criteria for the testing.

**2.** I chose a threshold value of 0.05 beacuse it means there is a 5% chance of incorrectly rejecting the null hypothesis when it's actually true. If the threshold value is set lower, 0.02 for example, the risk of false positives is reduced, but the "power" of the test to detect a real effect is also less. 


</div>

<div class="alert alert-info" style="border-radius: 15px; box-shadow: 4px 4px 4px; border: 1px solid ">

<b>Testing the hypothesis:</b> 

<b>Null Hypothesis H0</b> = The average user scores for Xbox One and PC platforms are the same.

<b>Alternative Hypothesis H1</b> = The average user scores for Xbox One and PC platforms are different.

</div>

```python
vgames_relevant['user_score'] = pd.to_numeric(vgames_relevant['user_score'], errors='coerce')

avg_xone_user_ratings = vgames_relevant.loc[vgames_relevant['platform'] == 'XOne', 'user_score']
avg_pc_user_ratings = vgames_relevant.loc[vgames_relevant['platform'] == 'PC', 'user_score']
   
alpha = 0.05
results = st.ttest_ind(avg_xone_user_ratings, avg_pc_user_ratings, nan_policy='omit')
   
print('p-value:', results.pvalue.round(4))
if results.pvalue < alpha:
       print("We reject the null hypothesis")
else:
       print("We can't reject the null hypothesis")   
```
